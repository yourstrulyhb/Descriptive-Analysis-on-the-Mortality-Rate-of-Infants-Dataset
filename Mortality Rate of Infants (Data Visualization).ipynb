### Plotting function

```python
def plot_mortality_rates(country_name='Australia'): # Case sensitive
    mortality_rate_country_z = mortality_rates_per_country.loc[mortality_rates_per_country["Country Name"] == country_name]
   
    if (not mortality_rate_country_z.empty):
        # Dropping qualitative columns
        mortality_rate_country_z = mortality_rate_country_z.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
        
#         pd.set_option("display.max_rows", None, "display.max_columns", None)
#         print(mortality_rate_country_z)
        
        # Plotting figures
        plt.figure(figsize=(45,20))
        plt.title(f'Mortality Rate of in Infants in the {country_name.capitalize()} from 1960-2021', fontsize=20, color='blue')
        plt.ylabel('Mortality Rate', fontsize=20)
        plt.xlabel('Year', fontsize=20)
        plt.plot(years, mortality_rate_country_z.values.tolist()[0])
        plt.grid()
        plt.show()
    
    else:
        return "No record found."
```

<h3 style='color: red'> Insert country name </h3>

```python
country_name = 'Philippines' #Replace country name between apostrophes
plot_mortality_rates(country_name) 
```

### Some insights on Mortality Rates of Infants in the Philippines

<ol>

### Getting some insights

```python
country_name = 'Philippines'
def get_insights(country_name):
    mortality_rate_country_z = mortality_rates_per_country.loc[mortality_rates_per_country["Country Name"] == country_name]
    
    # Dropping qualitative columns
    mortality_rate_country_z = mortality_rate_country_z.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
    
    max_val = mortality_rate_country_z.max(axis=1).values[0]
    idx_max = mortality_rate_country_z.idxmax(axis=1).values[0]
    min_val = mortality_rate_country_z.min(axis=1).values[0]
    idx_min = mortality_rate_country_z.idxmin(axis=1).values[0]
    
    print(f'The HIGHEST mortality rate of infants in {country_name} (per 1000 live births) was from the year {idx_max} with {max_val}.')
    print(f'The LOWEST mortality rate of infants in {country_name} (per 1000 live births) was from the year {idx_min} with {min_val}.')
    
get_insights(country_name)
```
