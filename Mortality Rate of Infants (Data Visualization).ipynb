# Mortality rate of infants in a country (1960-2021)

```python
mortality_rates_per_country.columns # Column names in the dataframe
years = mortality_rates_per_country.columns[4:].to_list() # Name of years of recorded 

print(f'{years[0]} - {years[-1]}') # We have records from year 1960-2021
```

### List of countries

```python
countries = mortality_rates_per_country['Country Name'].values.tolist()
countries
```

### Plotting function

```python
def plot_mortality_rates(country_name='Australia'): # Case sensitive
    mortality_rate_country_z = mortality_rates_per_country.loc[mortality_rates_per_country["Country Name"] == country_name]
   
    if (not mortality_rate_country_z.empty):
        # Dropping qualitative columns
        mortality_rate_country_z = mortality_rate_country_z.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
        
#         pd.set_option("display.max_rows", None, "display.max_columns", None)
#         print(mortality_rate_country_z)
        
        # Plotting figures
        plt.figure(figsize=(45,20))
        plt.title(f'Mortality rate of {country_name} from 1960-2021', fontsize=20, color='blue')
        plt.ylabel('Mortality Rate', fontsize=20)
        plt.xlabel('Year', fontsize=20)
        plt.plot(years, mortality_rate_country_z.values.tolist()[0])
        plt.grid()
        plt.show()
    
    else:
        return "No record found."
```

### `Insert country name`

```python
plot_mortality_rates('Philippines')
```
